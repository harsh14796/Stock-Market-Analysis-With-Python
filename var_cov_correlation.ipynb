```python
import pandas as pd
import numpy as np
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['PG','MSFT']

sec_data = pd.DataFrame()

for t in tickers :
    sec_data[t] = wb.DataReader(t,'quandl','2007-1-1','2018-3-1')['AdjClose']
```

```python
sec_data.info()
```

```python
sec_data = sec_data.iloc[::-1]
sec_returns = np.log(sec_data / sec_data.shift(1))
sec_returns
```

```python
PG_var = sec_returns['PG'].var() * 250
MSFT_var = sec_returns['MSFT'].var() * 250
print('P&G : ' + str(PG_var) + '\nMicrosoft : '+ str(MSFT_var))
```

```python
cov_matrix = sec_returns.cov() *250
cov_matrix
```

```python
corr_matrix = sec_returns.corr()
corr_matrix
```
